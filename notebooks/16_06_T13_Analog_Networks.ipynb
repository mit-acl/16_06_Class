# Lead 

```python
cct_lead = Circuit("""
P1 1 0 ; down=1.5, v_=v_i(t)
W1 1 2; right=1
R1 2 3 R; right=2
W22 2 22; up=1
C1 22 33 C; right=2
W33 33 3; down=1
R2 3 0_3 0.1*R; down=1.5
W03 0 0_3; right=1.5
W3 3 4; right=1
W04 0_3 0_4; right=1
P2 4 0_4; down, v^=v_o(t)
""")
cct8 = cct_lead.subs({'R': RR, 'C':CC})
cct8.draw()
```

```python
cct_lead.P1.transfer('P2')
```

```python
fv = np.logspace(0, 4, 1000)
w = fv *2*np.pi

Hl = cct8.P1.transfer('P2').limit('A', oo)
Num,Den = N(Hl).as_N_D()
NN = [Num.coeffs()[x].fval for x in range(len(Num.coeffs()))]
DD = [Den.coeffs()[x].fval for x in range(len(Den.coeffs()))]

G = ct.tf(np.real(NN),np.real(DD))
print("Poles = "+', '.join('{:6.2f}'.format(x) for x in G.poles()/2/np.pi))
print("Zeros = "+', '.join('{:6.2f}'.format(x) for x in G.zeros()/2/np.pi))
```

```python
Gf = G(complex(0,1)*w)

fig, ax = plt.subplots(2,1,figsize=(8, 5),dpi=150)
ax[0].loglog(fv,np.abs(Gf),color='b',ls='-')
ax[1].semilogx(fv,np.angle(Gf)*cm.r2d,color='b',ls='-')
ax[0].set_ylim([.05,1])
ax[1].set_ylim([0,90])
#ax[0].set_xlim([.1,10])
ax[0].set_ylabel("Magnitude")
ax[1].set_ylabel("Phase")
ax[1].set_xlabel("Frequency [Hz]")
ax[0].grid(True, which="both", ls="-",color='0.65')
ax[1].grid(True, which="both", ls="-",color='0.65')
plt.savefig('figs/Analog_Lead.pdf',dpi=600)
```

```python
CC = 0.1e-6 # capacitor value
RR = 10e3 # res value

cct_test1 = Circuit("""
P1 1 0_1; down
Rx 1 22; right
C 22 2; right
W 1 51; up 
W 2 52; up 
R1 51 52; right
R2 2_1 4_1; right
C2 4_1 3_1; right
W 4_1 53; up 
W 3_1 54; up 
Rf 53 54; right
E1 3_2 0_3 opamp 2_0 2 A; mirror
W 0_1 0; right
W 2_0 0; down
W 3_2 3; right
W 0 0_3; right
P2 3 0_3; down
W 2_1 2; down
W 3_1 3_2; down
""")
cct9 = cct_test1.subs({'Rx': RR, 'R1': RR, 'R2': 2.2*RR, 'Rf': 10*RR, 'C':CC/10, 'C1':CC, 'C2':CC})
cct9.draw()
```

```python
Hl = cct9.P1.transfer('P2').limit('A', oo)

lpoles = N(Hl.poles(1)).fval
cm.pretty_row_print([p/2/np.pi for p in lpoles],'Poles: ')
lzeros = N(Hl.zeros(1)).fval
cm.pretty_row_print([z/2/np.pi for z in lzeros],'Zeros: ')

Num,Den = N(Hl).as_N_D()
NN = [Num.coeffs()[x].fval for x in range(len(lzeros)+1)]
DD = [Den.coeffs()[x].fval for x in range(len(lpoles)+1)]
# rescale
MM = min(np.min(np.abs(NN)),np.min(np.abs(DD)))
NN = [x/MM for x in NN]
DD = [x/MM for x in DD]
G = ct.tf(NN,DD)
cm.show_tf_latex(G,name="G")
```

```python
fv = np.logspace(0, 4, 1000)
Gf = G(complex(0,1)*fv*2*np.pi)

fig, ax = plt.subplots(2,1,figsize=(8, 5),dpi=150)
ax[0].loglog(fv,np.abs(Gf),color='b',ls='-')
ax[1].semilogx(fv,np.unwrap(np.angle(Gf)*cm.r2d),color='b',ls='-')
#ax[0].set_ylim([.05,2])
ax[1].set_ylim([135,225])
ax[0].set_ylabel("Magnitude")
ax[1].set_ylabel("Phase")
ax[1].set_xlabel("Frequency [Hz]")
for ii in range(2):
    [ax[ii].axvline(x=np.abs(p/2/np.pi),c='r',ls='--') for p in lpoles]
    [ax[ii].axvline(x=np.abs(z/2/np.pi),c='c',ls='--') for z in lzeros]
bm.nicegrid(ax)
plt.savefig('figs/Active_Lag_Lead.pdf',dpi=600)

data = cm.Read_data("data/Active_Lag_lead.csv",cols=[0,1,2])

fig, ax = plt.subplots(2,1,figsize=(8, 5),dpi=150)
ax[0].loglog(fv,np.abs(Gf),color='b',ls='-',label='model')
ax[0].loglog(data[:,0],data[:,1],color='r',ls='--',label='data')
ax[1].semilogx(fv,np.unwrap(np.angle(Gf)*cm.r2d),color='b',ls='-',lw=3,label='model')
ax[1].semilogx(data[:,0],data[:,2],color='r',ls='--',label='data')
#ax[0].set_ylim([.05,2])
ax[1].set_ylim([135,225])
ax[0].set_ylabel("Magnitude")
ax[1].set_ylabel("Phase")
ax[1].set_xlabel("Frequency [Hz]")
for ii in range(2):
    [ax[ii].axvline(x=np.abs(p/2/np.pi),c='r',ls='--') for p in lpoles]
    [ax[ii].axvline(x=np.abs(z/2/np.pi),c='c',ls='--') for z in lzeros]
bm.nicegrid(ax)
ax[0].legend()
ax[1].legend()
plt.savefig('figs/Active_Lag_Lead_with_data.pdf',dpi=600)
```

```python
CC1 = 1e-6
CC2 = 1e-9
R1 = 100e3
R2 = 10e3
R3 = 100e3
print('   alpha     beta      zeta      wn')
for alp in [0.1,1,10]:
    C1 = alp*CC1
    for beta in [0.1,1,10]:
        C2 = beta*CC2
        wn = np.sqrt(1/(C1*C2*R2*R3))
        zeta = (R1*R2 + R2*R3 + R1*R3)/(C1*R1*R2*R3)/2./wn
        print(""+', '.join('{:>8.3f}'.format(i) for i in [alp,beta,zeta,wn/2/np.pi]))
```

```python
import numpy as np

# Define the target zeta and wn
target_zeta = 0.03
target_wn = 100.0

# Initialize variables to store the best match
best_match = None
best_error = float('inf')

# Iterate over all combinations of resistors and capacitors
for R1 in bojack_resistors:
    for R2 in bojack_resistors:
        for R3 in bojack_resistors:
            for C1 in bojack_capacitors:
                for C2 in bojack_capacitors:
                    # Calculate wn and zeta for the current combination
                    wn = np.sqrt(1 / (C1 * C2 * R2 * R3))
                    zeta = (R1 * R2 + R2 * R3 + R1 * R3) / (C1 * R1 * R2 * R3) / (2 * wn)
                    
                    # Calculate the error
                    error = np.abs(zeta - target_zeta) + np.abs(wn - target_wn)
                    
                    # Update the best match if the current error is smaller
                    if error < best_error:
                        best_error = error
                        best_match = (R1, R2, R3, C1, C2, zeta, wn)

# Print the best match
print(f"Best match: R1={best_match[0]}, R2={best_match[1]}, R3={best_match[2]}, C1={best_match[3]}, C2={best_match[4]}")
print(f"zeta={best_match[5]:4.2f}, wn={best_match[6]:4.2f}, best_error={best_error:.4f}")
```
