## <a id='toc1_2_'></a>[Analyze impact of adding Integrator](#toc0_)

```python
G0 = ct.tf(1, (1,3,2))

# pole to the left
G1 = G0*ct.tf((0, 1), (1, 0))

fig, ax = plt.subplots(1,2,figsize=(8, 4),dpi=150,constrained_layout = True)
klist, break_info0 = cm.Root_Locus_gains(G0, verbose=True)
ct.rlocus(G0, gains=klist,  ax=ax[0])

klist, break_info1 = cm.Root_Locus_gains(G1, verbose=True)
ct.rlocus(G1, gains=klist,  ax=ax[1])

for ll in range(2):
    dim = 4
    ax[ll].set_xlim([-dim, dim]);ax[ll].set_ylim([-dim, dim])
    bm.nicegrid(ax[ll])
    cm.add_break_info(ax[ll], break_info1 if ll == 1 else break_info0)
    cm.new_pzmap(G1 if ll == 1 else G0, ax=ax[ll])

plt.savefig("./figs/L12_1.pdf", dpi=600)
plt.show()
```
