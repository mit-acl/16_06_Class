# <a id='toc2_'></a>[Combined estimator/regulator design for a simple system](#toc0_)
$$G= \frac{1}{(s^2+s+1)}$$


```python
# System G
a = np.array([[0, 1],[-1, -1]])
nx = np.shape(a)[0]
b = np.array([0, 1]).reshape(nx,1)
c = np.array([1, 0])
d = 0

pr = np.array([-4+4*j, -4-4*j])
pe = np.array([-10.001, -10])

k = ct.place(a,b,pr);
l = ct.place(a.transpose(),c.reshape(nx,1),pe)
l = l.reshape(nx,1)
#
# For state space for G_c(s)
ac = a-b*k-l*c
bc = l
cc = k
dc = 0

G = ct.ss(a,b,c,d)
Gc = ct.ss(ac,bc,cc,dc)
L = G*Gc
Ltf = ct.ss2tf(L)
S = 1/(1+Ltf)
w = np.logspace(-1,2,400)
wcs,_ = cm.find_wc(w,L,find_all=True)
wpis,_ = cm.find_wpi(w,L,find_all=True)

Gf = G(w*j)
Gcf = Gc(j*w)
Lf = L(j*w)
Sf = S(j*w)

fig, ax = plt.subplots(2,1,figsize=(8, 5),dpi=150,constrained_layout = True)
ax[0].loglog(w,np.abs(Gf),'b',label=r'G')
ax[1].semilogx(w,np.unwrap(np.angle(Gf))*cm.r2d,'b',label=r'G')
ax[0].loglog(w,np.abs(Gcf),'r',label=r'Gc')
ax[1].semilogx(w,np.unwrap(np.angle(Gcf))*cm.r2d,'r',label=r'Gc')
ax[1].set_xlabel('Freq [r/s]')
ax[0].set_ylabel('Mag')
ax[1].set_ylabel('Phase (degs)')
ax[0].grid(True, which="both", ls="-",color='0.65')
ax[1].grid(True, which="both", ls="-",color='0.65')
ax[0].legend()
ax[0].set_ylim(1e-3,100)
ymax = np.max([np.max(np.abs(Gf)),np.max(np.abs(Gcf))])
ymax = 10**np.ceil(np.log10(1.5*ymax))
ax[0].set_ylim(1e-3,ymax)
plt.savefig("figs/DOFB0.pdf", dpi=600)
plt.show()

fig, ax = plt.subplots(2,1,figsize=(8, 5),dpi=150,constrained_layout = True)
ax[0].loglog(w,np.abs(Lf),'b',label=r'L')
ax[1].semilogx(w,np.unwrap(np.angle(Lf))*cm.r2d,'b',label=r'L')
ax[1].set_xlabel('Freq [r/s]')
ax[0].set_ylabel('Mag')
ax[1].set_ylabel('Phase (degs)')
ax[0].grid(True, which="both", ls="-",color='0.65')
ax[1].grid(True, which="both", ls="-",color='0.65')
ax[0].axhline(y=1,c='k',ls='--')
ax[1].axhline(y=-180,c='k',ls='--')
ax[0].legend()
ax[0].set_ylim(1e-3,100)
draw_vlines(ax, wcs, color='g')
draw_vlines(ax, wpis, color='m')
plt.savefig("figs/DOFB1.pdf", dpi=600)
plt.show()

fig, ax = plt.subplots(1,figsize=(8, 5),dpi=150,constrained_layout = True)
ax.loglog(w,np.abs(Sf),'b',label=r'S')
ax.loglog(w,np.abs(Lf),'r',label=r'L')
ax.set_xlabel('Freq [r/s]')
ax.set_ylabel('Mag')
ax.grid(True, which="both", ls="-",color='0.65')
ax.axhline(y=1,c='k',ls='--')
ax.legend()
ax.set_ylim(1e-3,100)
plt.savefig("figs/DOFB11.pdf", dpi=600)
plt.show()
```

```python
#
# loop dynamics L = G Gc
#
al = np.vstack((np.hstack((a,b*cc)),np.hstack((np.zeros((nx,nx)),ac))))
bl = np.vstack((np.zeros((nx,1)), bc ))
cl = np.hstack((c.reshape(1,nx), np.zeros((1,nx)) ))
dl = 0
L = ct.ss2tf(al,bl,cl,dl)
#
# closed-loop dynamics
# unity gain wrapped around loop L
#
acl = al - bl*cl
bcl = bl 
ccl = cl
dcl = d
# scale closed-loop to get zero ct.ss error
N = np.linalg.inv(np.dot(ccl,np.dot(np.linalg.inv(-acl),bcl)))
Gcl = cm.near_zero(ct.ss2tf(acl,bcl*N,ccl,dcl*N))
scl = Gcl.poles()
Gclf = Gcl(j*w)

assert np.allclose(np.sort(np.abs(scl)), np.sort(np.abs(np.concatenate((pr, pe)))), atol=1e-2), "scl is not close to pr + pe"

fig, ax = plt.subplots(1,figsize=(5, 5),dpi=150,constrained_layout = True)
klist = cm.Root_Locus_gains(L)
rl = ct.rlocus(L, gains= klist, ax=ax,color='b')
plt.plot(np.real(scl),np.imag(scl),'cs')
ax.set_aspect('equal')
ax.set_xlim([-25, 5]);
ax.set_ylim([-15, 15])
bm.nicegrid(ax)
plt.savefig("figs/DOFB2.pdf", dpi=600)
ax.set_xlabel('Real Axis')
ax.set_ylabel('Imag Axis')
plt.show()

fig, ax = plt.subplots(2,1,figsize=(8, 5),dpi=150,constrained_layout = True)
ax[0].loglog(w,np.abs(Gf),'b',label=r'G')
ax[1].semilogx(w,np.unwrap(np.angle(Gf))*cm.r2d,'b',label=r'G')
ax[0].loglog(w,np.abs(Gclf),'r',label=r'Gcl')
ax[1].semilogx(w,np.unwrap(np.angle(Gclf))*cm.r2d,'r',label=r'Gcl')
ax[1].set_xlabel('Freq [r/s]')
ax[0].set_ylabel('Mag')
ax[1].set_ylabel('Phase (degs)')
ax[0].grid(True, which="both", ls="-",color='0.65')
ax[1].grid(True, which="both", ls="-",color='0.65')
ax[0].legend()
ax[0].set_ylim(1e-2,10)
plt.savefig("figs/DOFB3.pdf", dpi=600)
plt.show()
```

```python
def DOFB(G,pr,pe,file_name='temp',w=np.logspace(-1,2,400),use_y_lims=True):
    '''
    DOFB(G,pr,pe,file_name='temp',w=np.logspace(-1,2,400),use_y_lims=True)
    Design a state feedback controller for the system G(s) using pole ct.placement.
    
    Inputs:
    G is a state space system
    pr is the desired closed-loop poles
    pe is the desired estimator poles
    file_name is the name of the file to save the figures
    w is the frequency vector for the Bode plots
    use_y_lims is a boolean to set the y-limits of the Bode plots

    Returns: 
    
    the state feedback gain k

    The function computes the state feedback gain k, the feedforward gain Nbar, and the closed-loop system Gcl.
    '''
    nx = G.A.shape[0]
    k = ct.place(G.A,G.B,pr)
    l = ct.place(G.A.T, G.C.reshape(nx,1), pe).T    
    #
    # For state space for G_c(s)
    #
    Gc = ct.ss(G.A-G.B*k-l*G.C,l,k,0)
    L = G*Gc
    L = cm.near_zero(ct.ss2tf(L.A,L.B,L.C,L.D))
    S = 1/(1+L)

    wcs, _ = cm.find_wc(w, L, find_all=True)
    wpis, _ = cm.find_wpi(w, L, find_all=True)

    Gf = G(w*j)
    Gcf = Gc(j*w)
    #
    # loop dynamics L = G Gc
    #
    al = np.vstack((np.hstack((G.A,G.B*Gc.C)),np.hstack((np.zeros((nx,nx)),Gc.A))))
    bl = np.vstack((np.zeros((nx,1)), Gc.B ))
    cl = np.hstack((G.C.reshape(1,nx), np.zeros((1,nx)) ))
    dl = 0
    L = ct.ss2tf(al,bl,cl,dl)
    Lf = L(j*w)
    #
    # closed-loop dynamics
    # unity gain wrapped around loop L
    #
    Gcl = ct.ss(al - bl*cl,bl,cl,d)
    # scale closed-loop to get zero ct.ss error
    N = np.linalg.inv(np.dot(Gcl.C,np.dot(np.linalg.inv(-Gcl.A),Gcl.B)))
    Gcl = Gcl*N
    scl = Gcl.poles()
    Gclf = Gcl(j*w)
    Sf = S(j*w)
    
    assert np.allclose(np.sort(np.abs(scl)), np.sort(np.abs(np.concatenate((pr, pe)))), atol=1e-2), "scl is not close to pr + pe"

    fig, ax = plt.subplots(2,1,figsize=(8, 5),dpi=150,constrained_layout = True)
    ax[0].loglog(w,np.abs(Gf),'b',label=r'G')
    ax[1].semilogx(w,np.unwrap(np.angle(Gf))*cm.r2d,'b',label=r'G')
    ax[0].loglog(w,np.abs(Gcf),'r',label=r'Gc')
    ax[1].semilogx(w,np.unwrap(np.angle(Gcf))*cm.r2d,'r',label=r'Gc')
    ax[1].set_xlabel('Freq [r/s]')
    ax[0].set_ylabel('Mag')
    ax[1].set_ylabel('Phase (degs)')
    ax[0].grid(True, which="both", ls="-",color='0.65')
    ax[1].grid(True, which="both", ls="-",color='0.65')
    ax[0].legend()
    if use_y_lims:
        ymax = np.max((np.max(np.abs(Gf)),np.max(np.abs(Gcf))))
        ymax = 10**np.ceil(np.log10(1.5*ymax))
        ax[0].set_ylim(1e-3,ymax)

    plt.savefig("figs/DOFB"+file_name+"_0.pdf", dpi=600)
    plt.show()

    fig, ax = plt.subplots(2,1,figsize=(8, 5),dpi=150,constrained_layout = True)
    ax[0].loglog(w,np.abs(Lf),'b',label=r'L')
    ax[1].semilogx(w,cm.pshift(np.unwrap(np.angle(Lf)))*cm.r2d,'b',label=r'L')
    test_val = np.mean(cm.pshift(np.unwrap(np.angle(Lf)))*cm.r2d)
    ax[1].set_xlabel('Freq [r/s]')
    ax[0].set_ylabel('Mag')
    ax[1].set_ylabel('Phase (degs)')
    ax[0].grid(True, which="both", ls="-",color='0.65')
    ax[1].grid(True, which="both", ls="-",color='0.65')
    ax[0].axhline(y=1,c='k',ls='--')
    if test_val < 0:
        ax[1].axhline(y=-180,c='k',ls='--')
    else:
        ax[1].axhline(y=180,c='k',ls='--')
    ax[0].legend()
    if use_y_lims:
        ymax = np.max((np.max(np.abs(Lf)),np.max(np.abs(Lf))))
        ymax = 10**np.ceil(np.log10(1.5*ymax))
        ax[0].set_ylim(1e-3,ymax)

    draw_vlines(ax, wcs, color='g')
    draw_vlines(ax, wpis, color='m')
    plt.savefig("figs/DOFB"+file_name+"_1.pdf", dpi=600)
    plt.show()

    if 0:
        fig, ax = plt.subplots(1,figsize=(5, 5),dpi=150,constrained_layout = True)
        klist = cm.Root_Locus_gains(L)
        rl = ct.rlocus(L, gains=klist, grid=False,ax=ax,color='b')
        plt.plot(np.real(scl),np.imag(scl),'cs')
        ax.set_aspect('equal')
        ax.set_xlim([-25, 5]);
        ax.set_ylim([-15, 15])
        bm.nicegrid(ax)
        ax.set_xlabel('Real Axis')
        ax.set_ylabel('Imag Axis')
        plt.savefig("figs/DOFB"+file_name+"_2.pdf", dpi=600)
        plt.show()

    fig, ax = plt.subplots(2,1,figsize=(8, 5),dpi=150,constrained_layout = True)
    ax[0].loglog(w,np.abs(Gf),'b',label=r'G')
    ax[1].semilogx(w,np.unwrap(np.angle(Gf))*cm.r2d,'b',label=r'G')
    ax[0].loglog(w,np.abs(Gclf),'r',label=r'Gcl')
    ax[1].semilogx(w,np.unwrap(np.angle(Gclf))*cm.r2d,'r',label=r'Gcl')
    ax[1].set_xlabel('Freq [r/s]')
    ax[0].set_ylabel('Mag')
    ax[1].set_ylabel('Phase (degs)')
    ax[0].grid(True, which="both", ls="-",color='0.65')
    ax[1].grid(True, which="both", ls="-",color='0.65')
    ax[0].legend()
    if use_y_lims:
        ymax = np.max((np.max(np.abs(Gf)),np.max(np.abs(Gclf))))
        ymax = 10**np.ceil(np.log10(1.5*ymax))
        ax[0].set_ylim(1e-3,ymax)
        
    plt.savefig("figs/DOFB"+file_name+"_3.pdf", dpi=600)
    plt.show()
    
    fig, ax = plt.subplots(1,figsize=(8, 5),dpi=150,constrained_layout = True)
    ax.loglog(w,np.abs(Sf),'b',label=r'S')
    ax.loglog(w,np.abs(Lf),'r',label=r'L')
    ax.axhline(y=1,c='salmon',ls='--',zorder=0)
    ax.set_xlabel('Freq [r/s]')
    ax.set_ylabel('Mag')
    ax.grid(True, which="both", ls="-",color='0.65')
    ax.legend()
    if use_y_lims:
        smax = np.max(np.abs(Sf))
        smax = 10**np.ceil(np.log10(smax))
        ax.set_ylim(1e-2,smax)
        
    plt.savefig("figs/DOFB"+file_name+"_4.pdf", dpi=600)
    plt.show()

    return Gc,Gcl,Ltf
```

## <a id='toc2_1_'></a>[DOFB Example 1](#toc0_)

```python
G1 = ct.ss(ct.tf(8*14*20,np.convolve([1, 8],np.convolve([1, 14],[1, 20]))))
file_name = 'examp1'
nx = len(G1.poles())
zz,pp,kk = butter(nx,20,output='zpk',analog=True)
pe = [2*p for p in pp]
Gc1,Gcl1,_ = DOFB(G1,pp,pe,file_name=file_name, w=np.logspace(0,3,400))
cm.pretty_row_print(Gcl1.poles(),"Closed-loop poles: ")
```
